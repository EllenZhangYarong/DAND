```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_city = pd.read_csv("./dataset/city_data.csv")
df_city.head()
```

```python
len(df_city)
```

```python
df_city_temp = df_city["avg_temp"]
df_city_temp.head()
```

```python
df_city_5_ma = df_city_temp.rolling(10).mean()
```

```python
df_global = pd.read_csv("./dataset/global_data.csv")
df_global.head()
```

```python
df_oslo = pd.read_csv("./dataset/oslo.csv")
df_oslo = df_oslo.dropna()
df_oslo.head()
```

```python
len(df_oslo)
```

```python
plt.figure(figsize=(15,10))
x = np.linspace(100, 260, len(df_city))
plt.grid(True)
# plt.plot(df_city_5_ma, label='Wellinton Weather Changing')
plt.plot(x, df_city["avg_temp"].rolling(50).mean(), label='Wellington')
plt.plot(df_global["avg_temp"].rolling(50).mean(), label='Global')
plt.plot(df_oslo["avg_temp"].rolling(50).mean(), label='Oslo')

plt.xlabel('Time(in 50 years)')
plt.ylabel('Average Tempreture')

plt.legend(loc=2)

plt.savefig('./pics/weather_50.png')
```

```python
plt.figure(figsize=(15,10))
x = np.linspace(100, 260, len(df_city))
plt.grid(True)
# plt.plot(df_city_5_ma, label='Wellinton Weather Changing')
plt.plot(x, df_city["avg_temp"].rolling(20).std(), label='Wellington')
plt.plot(df_global["avg_temp"].rolling(20).std(), label='Global')
plt.plot(df_oslo["avg_temp"].rolling(20).std(), label='Oslo')
plt.legend(loc=2)
```
